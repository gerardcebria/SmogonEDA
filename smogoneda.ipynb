```python
import requests
import urllib.request
import re
import pandas as pd
import numpy as np
```

```python
def getTotalBattles(text): 
    line_str = text.decode()
    matches = re.findall(r'\d+', line_str)
    numbers = [int(match) for match in matches]
    return numbers

def getValuesFromRow(text):
    return text.decode().replace(" ", "").replace("\t", "").replace("\n", "").split("|")[1:-1]
    
def appendData(text):
    # Define an array to append to the DataFrame
    values = getValuesFromRow(text)
    if len(values)>0:
        new_data = np.array([values])
        new_df = pd.DataFrame(new_data, columns = ['Rank', 'Pokemon', 'Usage%', 'Raw', '%', 'Real', '%'])
        return new_df
```

```python
def getDataFromUrl(url, headers):
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an exception for 4xx and 5xx status codes

        json_data = response.json()
        return json_data

    except requests.exceptions.RequestException as e:
        print(f"Error occurred during the request: {e}")
        return None
```

```python
def get_player_info(player_info):
    player = {
        "player_name": player_info[3],
        "player_rank": player_info[5],
    }
    return player


def get_team_info(team_info):
    team = {
        "player": team_info[2],
        "poke_name": team_info[3].split(",")[0],
        "poke_lvl": team_info[3].split(",")[1],
        "poke_gender": 'X' if len(team_info[3].split(",")) <=2 else team_info[3].split(",")[2] 
    }
    return team

def get_start(start_info):
    start = {
        "position": start_info[2].split(':')[0],
        "poke": start_info[2].split(':')[1]
    }
    return start

def get_turn_info(n_turn, turn_info):
    return turn_info

def get_move_info(n_turn, turn_info):
    turn = {
        "n_turn": n_turn,
        "source": turn_info[2],
        "tera": None,
        "move": turn_info[3],
        "objective": turn_info[4],
        "effect": '',
        "result": None,
        "dmg_qty": None,
        "stat_qty": None,        
    }
    return turn
    
def get_efect_info(move, effect_info):
    move["effect"] = effect_info[1]
    move["result"] = effect_info[2]
    move["dmg_qty"] = None if len(effect_info) <= 3 else effect_info[3]
    move["stat_qty"] = None if len(effect_info) <= 4 else effect_info[4]
    return move

def parse_battle_string(battle_string):


# Create an empty DataFrame with the specified columns
    moves = pd.DataFrame(columns=["n_turn", "source", "tera", "move", "objective", "effect", "result","dmg_qty", "stat_qty"])
    lines = battle_string.strip().split("\n")
    players = []
    teams = []
    start = []
    n_turn = 0
    turns = []
    move = {}

    for line in lines:
        if line.startswith("|player|"):
            players.append(get_player_info(line.split("|")))
        elif line.startswith("|poke|"):
            teams.append(get_team_info(line.split("|")))
        elif line.startswith("|switch|") and len(start) < 4:
            start.append(get_start(line.split("|")))
        elif line.startswith("|turn|"):
            if n_turn != 0:
                turns.append(get_turn_info(n_turn, line.split("|")))
            n_turn = int(line.split("|")[2])
            
        elif n_turn > 0 and 'inactive' not in line.split("|"):
            if 'move' in line.split("|"):
                move = get_move_info(n_turn, line.split("|"))
            elif any('-' in item for item in line.split("|")):
                effect = get_efect_info(move, line.split("|"))
                moves = moves.append(effect, ignore_index = True)
        
    return players, teams, start, turns, moves

players, teams, start, turns, moves_df = parse_battle_string(lines)
moves_df
```
